```python
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np
```

```python
df = pd.read_csv("raw.csv")
```

```python
xCol= df.columns[2:]
xData = df[xCol]
yData = df["target"]
```

```python
sbn.boxplot(x= 'variable', y ='value', data = pd.melt(xData))
plt.show()
```

```python
yData.value_counts()
```

```python
threshold =.5
absoluteCorAboveThresh = np.absolute(df.corr())>threshold
numberCorrelatedwithOthers =absoluteCorAboveThresh.apply(sum)
numberCorrelatedwithOthers.plot(kind= 'hist')
plt.show()
# Indicates that majority of the variables are not correlated with each other.
```
